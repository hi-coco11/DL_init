```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Data Generation

```python
##1. Generating data
center = np.array([[-2, -2], [2, 2], [2, -2]])
point_class = 100
m = point_class * len(center)

X = np.zeros((m, 2)) ## 좌표 저장
y = np.zeros(m, dtype=int) ## 라벨 저장

for i, center in enumerate(center):
    i_center_first = i * point_class
    i_center_last = (i+1) * point_class
    X[i_center_first:i_center_last] = center + np.random.randn(point_class, 2) ## 각 center에 노이즈 추가
    y[i_center_first:i_center_last] = i
```

```python
##2. Labelling
Xclass0 = X[y==0]
Xclass1 = X[y==1]
Xclass2 = X[y==2]
plt.figure(figsize=(8,6))
```

```python
##3. Visualization
plt.scatter(Xclass0[:,0], Xclass0[:,1], color='red', label = 'Class 0')
plt.scatter(Xclass1[:,0], Xclass1[:,1], color='blue', label = 'Class 1')
plt.scatter(Xclass2[:,0], Xclass2[:,1], color='green', label = 'Class 2')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('Generated Classification Dataset')
```

```python
X = np.hstack((X, np.ones((300,1))))
X = X.T ## (3, 300) // WX 형태

## 정답 라벨링
Y = np.zeros((3, m)) 
for i in range(y.shape[0]): 
    Y[y[i], i] = 1
```
